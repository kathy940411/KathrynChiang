<strong>Part 2: Deterministic</strong><br>
We clearly see with the plot an overall increasing trend from 1994 to 2005 regarding the amount of milk production. Apparently, an order one polynomial is good enough to summarize the trend. Next, we will fit the seasonal components to the data after the trend is removed. We will use a set of harmonic terms to model the seasonal components. The seasonal part is
</p>
\begin{equation}
S_t = a_0 + \sum_{j=1}^{k} (a_s \sin(2 \pi\lambda_j t) + a_c \cos(2\pi\lambda_j t))
\end{equation}

<p>
If we based on ACF and PACF, we will pick AR(1) model. However, the noise does not appear to be constant over time. Since ARMA model requires stationarity assumption on the series. Therefore, we cannot fit a model to the residuals directly.  Dickey-Fully Test is performed to check the stationarity assumption. The p-value is 0.0678. Therefore, we fail to reject the null hypothesis at 5% significance level. That means, the residuals are not stationary. A differencing may be performed to make the series look stationary. So we may need to perform differencing before we fit a time series model. 
</p>
<p>
<strong>Part 3: Time series model</strong><br>
After we do a non-seasonal differencing on the residuals, the result is given in the following figures. We can see the mean is now constant across time. However, the correlation is still dependent on time. So we may need to perform stationary differencing. It seems to be tedious for this project. So, we check for stationary again with Dickey-Fully Test first. Now, the new p-value is 0.01. Therefore, we reject the null hypothesis at 5% significance level. That means, we can assume the series is stationary. So, we may not need to do seasonal differencing.
</p>
<img src="http://i.imgur.com/fevT8Ga.png">
<p>
Now, we can fit a ARMA model to the series. Before we fit a model, let us check the ACF and PACF again. It is hard to tell the model. But it seems to be AR(4) from PACF. Next, we will search the best model throughout ARMA models up to p=5 and q=5 based on AIC criterion.
</p>
<img src="http://i.imgur.com/LGUxSV0.png">
<p>
The best model selected is AR(4) with estimated coefficients given below

<p>
Next, we perform Box-Ljung test on the residuals. The corresponding p-value is 0.5066. Therefore, we fail to reject the null hypothesis that the noises are independent. Therefore, we do not see any dependent structure in the residuals after we fitted the selected time series model. 
</p>
